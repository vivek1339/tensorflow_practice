```python
import datalab.bigquery as bq
import tensorflow as tf
import pandas as pd
import numpy as np
import shutil

print(tf.__version__)
```

```python
csv_coloumns=['height','radius','area']
features=csv_coloumns[0:len(csv_coloumns)-1]
label=csv_coloumns[len(csv_coloumns)-1]
df_train=pd.read_csv('./data.csv', names = csv_coloumns)
df_test=pd.read_csv('./dtaa.csv', names = csv_coloumns)
```

```python
def make_input_train(df,no_of_epochs):
    return tf.estimator.inputs.pandas_input_fn(
        x=df,
        y=df[label],
        batch_size=128,
        num_epochs=no_of_epochs,
        shuffle=True,
        num_threads=1
    )
def input_test():
    return (df_test.batch(128)
            .make_one_shot_iterator().get_next())
def make_prediction_input_fn(df, num_epochs):
  return tf.estimator.inputs.pandas_input_fn(
    x = df,
    y = None,
    batch_size = 128,
    num_epochs = num_epochs,
    shuffle = True,
    queue_capacity = 1000,
    num_threads = 1
  )
def make_feature_coloumns():
    input_coloumns=[tf.feature_column.numeric_column(k) for k in features]
    return input_coloumns
```

```python
print(features)
print(label)
print(make_feature_coloumns())
```

```python
model=tf.estimator.DNNRegressor(hidden_units=[10,10],feature_columns=make_feature_coloumns())
```

```python
for i in range(10):
    model.train(input_fn=make_input_train(df_train,no_of_epochs=100),steps=50000)
```

```python
def print_rmse(model, name, df):
  metrics = model.evaluate(input_fn = make_input_train(df, 1))
  print('RMSE on {} dataset = {}'.format(name, np.sqrt(metrics['average_loss'])))
print_rmse(model, 'validation', df_test)
```

```python
predictions = model.predict(input_fn = make_prediction_input_fn(df_test, 1))
for i in range(5):
  print(next(predictions))
```
